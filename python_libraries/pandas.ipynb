#### Merge, Join, Concatenate

```python
df1=pd.read_csv(r"C:\Users\Lenovo\Downloads\LOTR.csv")
```

```python
df1
```

```python
df2=pd.read_csv(r"C:\Users\Lenovo\Downloads\LOTR 2.csv")
```

```python
df2
```

```python
df1.merge(df2)    #here default the join is the inner join
```

```python
df1.merge(df2,how='left',on='FellowshipID')     #similar like left join in sql and the column on which it is done is FellowshipID
```

```python
df1.merge(df2,how='outer')   #Similar like outer join in sql
```

```python
df1.merge(df2,how='right')   #Right join
```

```python
df1.merge(df2,how='inner',on=['FellowshipID','FirstName'])   #Merging based on two columns
```

```python
df1.merge(df2,how='cross')    #This is like cartesian product. cross join
```

```python
df4= df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix='_Left',rsuffix='_right',how='outer')
df4
```

Difference between join and merge
* merge() works on 'Columns', whereas join() works on 'Indexes'
* In merge default join is 'Inner join", but in join() the default one is "Left join"
* Use merge() when you need more control over the join, especially when dealing with column-based joins.
* Use join() when working with DataFrames that are already indexed properly.

```python
pd.concat([df1,df2])   #Concatenating two columns
```

## Pandas Data cleaning

```python
df=pd.read_excel(r"C:\Users\Lenovo\Downloads\Customer Call List.xlsx")
```

```python
df
```

```python
df=df.drop_duplicates()   #Gropping duplicates
```

```python
df=df.drop(columns="Not_Useful_Column")   #Dropping columns
```

```python
df['Last_Name']   #Accessing last name column
```

```python
df
```

```python
# Cleaning the text data. stripping the data using lstrip,rstrp and strip methods
df['Last_Name']=df['Last_Name'].str.lstrip("...")
df['Last_Name']=df['Last_Name'].str.lstrip("/")
df['Last_Name']=df['Last_Name'].str.rstrip("_")
```

```python
df
```

```python
#pip install regex    #regular expression library
```

```python
import regex as re    #importing regular expression as re
#applying a regular expression to clean phone numbers column
df['Phone_Number'] = df['Phone_Number'].astype(str).apply(lambda x: re.sub(r'\D', '', x))
```

```python
df['Phone_Number']
```

```python
#modeling phone numbers column like "123-545-5421"
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-'+x[3:6]+'-'+x[6:10])
df
```

```python
#Splitting address column based on comma(,) 
df[['Street Address','State','Zip_code']]=df['Address'].str.split(',',n=2,expand=True)
#the parameter n=2 controls the maximum number of splits that will occur.
```

```python
df
```

```python
#Replacing the Yes, No column as Y,N
df['Paying Customer']=df['Paying Customer'].replace(['Yes','No'],['Y','N'])
```

```python
df
```

```python
df=df.fillna('')   #Filling null values with space
```

```python
#replacing the text data with other text data
df['Do_Not_Contact']=df['Do_Not_Contact'].replace(['Yes','No'],['Y','N'])
```

```python
df
```

```python
#Changing the dataframe and dropping the columns where the do not contact =='Y'
for x in df.index:
    if df.loc[x,'Do_Not_Contact']=='Y':
        df.drop(x,inplace=True)
df
```

```python
#Dropping the columns where the Phone_Number have space
for x in df.index:
    if df.loc[x,'Phone_Number']=='':
        df.drop(x,inplace=True)
df
```

```python
df['Phone_Number']=df['Phone_Number'].replace('--','') #Replacing the different value in phone number column
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('','N')  #Replacing the space in do not contact column with N
```

```python
import pandas as pd

# Creating a DataFrame
data = {
    'Region': ['North', 'South', 'North', 'East', 'South', 'East', 'North'],
    'Category': ['Electronics', 'Furniture', 'Electronics', 'Furniture', 'Electronics', 'Furniture', 'Furniture'],
    'Sales': [200, 150, 300, 100, 250, 180, 220]
}

dfpivot = pd.DataFrame(data)

print("Original DataFrame:")
print(dfpivot)
```

```python
# Creating a pivot table
pivot = dfpivot.pivot_table(values='Sales', index='Region', columns='Category', aggfunc='sum', fill_value=0)

print("\nPivot Table Output:")
print(pivot)
```

```python
# Unpivot (Melt) the pivot table
unpivoted = pivot.reset_index().melt(id_vars='Region', var_name='Category', value_name='Sales')

print("\nUnpivoted DataFrame:")
print(unpivoted)
```

```python
import pandas as pd

# Creating a sample DataFrame
data = {
    'CustomerID': [101, 102, 103],
    'First_Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}

dfeval = pd.DataFrame(data)

# Using eval() to calculate a new column
dfeval['New_Salary'] = dfeval.eval('Salary * 1.10')  # 10% increase in salary

# Using eval() to create a condition
dfeval['Is_Senior'] = dfeval.eval('Age > 30')  # Boolean column for seniority

print(dfeval)
```
